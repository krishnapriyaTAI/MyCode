```python
# Zomato Data Analysis using Pandas and Numpy
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataframe = pd.read_csv("Zomato-data-.csv")
dataframe
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'],color = 'Green',width =0.5)
plt.xlabel("Type of restaurant")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='Blue', marker='X')
plt.xlabel('Type of restaurant', c='Green', size=20)
plt.ylabel('Votes', c='Green', size=20)
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']
print('Restaurant(s) with the maximum votes:')
restaurant_with_max_votes
```

```python
sns.countplot(x=dataframe['online_order'],color='green',width = 0.3)
```

```python
plt.hist(dataframe['rate'],bins=5,color="indigo")
plt.title('Ratings Distribution',color="Green")
plt.show()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data,color="orange",width=0.9)
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe,color="cyan")
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap',color = "Green")
plt.xlabel('Online Order',color="Blue")
plt.ylabel('Listed In (Type)',color="Blue")
plt.show()
```
